```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, f1_score, accuracy_score, roc_auc_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression

# Load data
trainingData = pd.read_csv("amazon_train.csv")
testingData = pd.read_csv("amazon_test.csv")

# Define target variable
y = trainingData["overall"]

# Create TF-IDF vectorizer
myVect = TfidfVectorizer()
myVect.fit(trainingData["reviewText"].fillna('') + ' ' + trainingData["summary"].fillna(''))

# Transform data
xTest = myVect.transform(testingData['reviewText'].fillna('') + ' ' + testingData['summary'].fillna(''))
xTrain = myVect.transform(trainingData['reviewText'].fillna('') + ' ' + trainingData['summary'].fillna(''))

# Split data
x_train_split, x_validation, y_train_split, y_validation = train_test_split(xTrain, y, test_size=0.2, random_state=9)

# Dictionary for potential hyperparameters for each model
modelsParameters = {
    'Multinomial Naive Bayes': {
        'model': MultinomialNB(),
        'params': {
            'alpha': np.linspace(0.1, 1.0, num=2),
            'fit_prior': [True]
        },
    },
    'Random Forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [200],
            'max_depth': [20],
        }
    }
}

# Function to create predictions file
def createFile(model, model_name, test_ids, x_test_data):
    y_sub = model.predict(x_test_data)
    submiss = pd.DataFrame({'pred': y_sub, 'id': test_ids})
    fileName = f"{model_name} Predictions.csv"
    submiss.to_csv(fileName, index=False)
    print(fileName)

# Loop through models
for model_name, model_info in modelsParameters.items():
    print(f"Now doing {model_name}...")
    grid_search = GridSearchCV(model_info['model'], model_info['params'], cv=5, scoring="f1_macro")
    grid_search.fit(x_train_split, y_train_split)

    bestParams = grid_search.best_params_
    print(f"Ideal Params {model_name}: {bestParams}")

    trained_model = model_info['model'].set_params(**bestParams)
    trained_model.fit(x_train_split, y_train_split)
    y_predicted = trained_model.predict(x_validation)

    # Confusion Matrix
    cfu = confusion_matrix(y_validation, y_predicted)
    print(f"{model_name} \nConfusion Matrix:")
    print(cfu)

    # F1 score
    f1_s = f1_score(y_validation, y_predicted, average="macro")
    print(f"F1 score: {f1_s:.2f}")

    # Accuracy Score
    a_s = accuracy_score(y_validation, y_predicted)
    print(f"Accuracy Score: {a_s:.2f}")

    # ROC-AUC Score (for each class against the rest)
    y_proba = trained_model.predict_proba(x_validation)
    roc_auc_scores = []
    for i in range(len(trained_model.classes_)):
        roc_auc_i = roc_auc_score((y_validation == trained_model.classes_[i]).astype(int), y_proba[:, i])
        roc_auc_scores.append(roc_auc_i)
        print(f"ROC-AUC Score (Class {trained_model.classes_[i]}): {roc_auc_i:.2f}")

    print("\n")

# Create predictions for each model
testingIDS = testingData["id"]
for model_name, model_info in modelsParameters.items():
    print(f"{model_name} now being made")
    trained_model = model_info['model']
    createFile(trained_model, model_name, testingIDS, xTest)

```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import VotingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
from sklearn.metrics import confusion_matrix
# Load data
trainingData = pd.read_csv("amazon_train.csv")
testingData = pd.read_csv("amazon_test.csv")

# Combine 'reviewText' and 'summary'
trainingData['text_combined'] = trainingData['reviewText'].fillna('') + ' ' + trainingData['summary'].fillna('')
testingData['text_combined'] = testingData['reviewText'].fillna('') + ' ' + testingData['summary'].fillna('')

# Vectorize 'text_combined'
tfidf_vectorizer = TfidfVectorizer()
xTrain_tfidf = tfidf_vectorizer.fit_transform(trainingData['text_combined'])
xTest_tfidf = tfidf_vectorizer.transform(testingData['text_combined'])

# Split the data
y = trainingData["overall"]
x_train_split, x_validation, y_train_split, y_validation = train_test_split(xTrain_tfidf, y, test_size=0.10, random_state=9)


# Define classifiers
classifier1 = MultinomialNB(alpha=1.67, fit_prior=True)
classifier2 = RandomForestClassifier(n_estimators=300, class_weight="balanced", max_depth=20)
classifier3 = LogisticRegression(max_iter=3000, multi_class='multinomial', solver='lbfgs')


# Create an ensemble of classifiers
ensemble_classifier = VotingClassifier(estimators=[
    ('nb', classifier1),
    ('rf', classifier2),
    ('lr', classifier3)
], voting='soft', weights=[1, 2, 1])

# Train the ensemble classifier
ensemble_classifier.fit(x_train_split, y_train_split)

# Make predictions on the validation set
y_predicted = ensemble_classifier.predict(x_validation)

# Evaluate the model
cfu = confusion_matrix(y_validation, y_predicted)
f1_s = f1_score(y_validation, y_predicted, average="macro")
a_s = accuracy_score(y_validation, y_predicted)

print("Ensemble Classifier \nConfusion Matrix")
print(cfu)
print(f"F1 score: {f1_s:.3f}")
print(f"Accuracy Score: {a_s:.3f}")

# Make predictions on the test set
y_test_predicted = ensemble_classifier.predict(xTest_tfidf)

# Create submission file
submission = pd.DataFrame({'pred': y_test_predicted, 'id': testingData["id"]})
submission.to_csv("Ensemble_Predictions.csv", index=False)
print("Ensemble Classifier predictions on the test set saved to 'Ensemble_Predictions.csv'")
```
